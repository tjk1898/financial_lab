# 简单计算股票连续上涨的概率
1. 获取所有股票近一个月的K线
2. 统计在出现一个暴涨（10%）后，再次出现暴涨（10%）的概率。
   1. 获取第一次出现暴涨（10%）的次数
   2. 获取连续出现两个暴涨的次数

```python
# 加载csv
# 遍历code
# 找到第一次出现暴涨（7%）的日期，进行统计，first_rise_count，并记录股票的代码和日期
# 找到第二次出现暴涨（7%）的日期，进行统计，second_rise_count，并记录股票的代码和日期
# 找到第二次出现暴涨（7%）的日期，进行统计，second_slight_rise_count，并记录股票的代码和日期
# 第二天出现下跌的日期，进行统计，second_decline_count，并记录股票的代码和日期
#
# 通过 second_rise_count/first_rise_count 可以得出连续暴涨的概率

def add_before_rate(df):
    """
    计算当前增长率和前一天增长率
    :param df:
    :return:
    """
    tmp_df = pd.DataFrame()
    # 获得codes
    codes = df['code'].unique()
    for code in codes:
        # 取单只股票
        single_stock = df[df['code'] == code]
        # 重新index
        single_stock = single_stock.reset_index(drop=False)
        single_stock = single_stock.drop(columns=['index'])
        # 添加增长率
        single_stock['rise_rate'] = (single_stock['close'] - single_stock['open']) / single_stock['open']
        # 添加前一天的增长率
        last_day = single_stock.drop(len(single_stock) - 1, axis=0)
        last_day.index += 1
        single_stock['before_day_rate'] = last_day['rise_rate']
        tmp_df = tmp_df.append(single_stock)
    tmp_df = tmp_df.reset_index(drop=False)
    tmp_df = tmp_df.drop(columns=['index'])
    return tmp_df
```

```python
tmp_df = pd.read_csv('/Users/jinkun.tian/gitee/financial_lab/AKshareTest/source/stock_price_20220401_0426.csv')
data_df = add_before_rate(tmp_df)
data_df.to_csv('/Users/jinkun.tian/gitee/financial_lab/AKshareTest/source/tmp_stock_add_rate.csv')
continue_rise=data_df[(data_df['rise_rate']>0.1) & (data_df['before_day_rate']>0.1)][['code','date','rise_rate']]
first_rise=data_df[(data_df['rise_rate']>0.1) & (data_df['before_day_rate']<0)][['code','date','rise_rate']]
```

```python
print("连续两天涨停的股票",len(continue_rise))
print("前一天下跌，第二天涨停的股票",len(first_rise))
```

```python
first_rise_second_declines=data_df[(data_df['rise_rate']<0) & (data_df['before_day_rate']>0.1)][['code','date','rise_rate']]
print("前一天涨停,第二天下跌的股票",len(first_rise_second_declines))
```
