```python
import os;
```

```python
current_dir = os.getcwd()
data_path = os.path.join(current_dir, 'data')
parent_path = os.path.dirname(current_dir)
```

```python
list = os.listdir(data_path)
file_map = {};
for filename in list:
    code = filename.replace(".csv", "")[2:]
    file_map[code] = filename
```

```python
import datetime
def add_date(date: str = "2022-05-01", delta: int = 1):
    dateTime_p = datetime.datetime.strptime(date, '%Y-%m-%d')
    return (dateTime_p + datetime.timedelta(days=delta)).strftime('%Y-%m-%d')
```

```python
import datetime

today = datetime.date.today()
formatted_today = today.strftime('%Y-%m-%d')
for code in qczc_df["代码"]:
    print(code)
    try:
        file_name = file_map[code]
    except:
        print(code + " is none")
        continue
    tmp_df = pd.read_csv(data_path + "/" + file_name)
    tmp_df = tmp_df.loc[:, ~tmp_df.columns.str.contains("^Unnamed")]
    start_time = tmp_df.iloc[-1]['date']
    start_time = add_date(start_time, 1)
    df = ak.stock_zh_a_daily(file_name.replace(".csv", ""), start_time, formatted_today, adjust="qfq")
    df['amount'] = df['volume']/100
    df[['amount']] = df[['amount']].astype('int')
    df['code'] = tmp_df.iloc[-1]['code']
    df['name'] = tmp_df.iloc[-1]['name']
    tmp_df = tmp_df.append(df[['date', 'open', 'close', 'high', 'low', 'amount', 'code', 'name']], ignore_index = True)
    tmp_df.to_csv(data_path+"/"+file_name, index=None)
    print(file_name+" updated.")
```

```python
def update_csv(file_map,df):
    today = datetime.date.today()
    formatted_today = today.strftime('%Y-%m-%d')
    for code in df["代码"]:
        print(code)
        try:
            file_name = file_map[code]
        except:
            print(code + " is none")
            continue
        tmp_df = pd.read_csv(data_path + "/" + file_name)
        tmp_df = tmp_df.loc[:, ~tmp_df.columns.str.contains("^Unnamed")]
        start_time = tmp_df.iloc[-1]['date']
        start_time = add_date(start_time, 1)
        st = datetime.datetime.strptime(start_time,'%Y-%m-%d')
        et = datetime.datetime.strptime(formatted_today,'%Y-%m-%d')
        if st>=et:
            continue
        df = ak.stock_zh_a_daily(file_name.replace(".csv", ""), start_time, formatted_today, adjust="qfq")
        df['amount'] = df['volume']/100
        df[['amount']] = df[['amount']].astype('int')
        df['code'] = tmp_df.iloc[-1]['code']
        df['name'] = tmp_df.iloc[-1]['name']
        tmp_df = tmp_df.append(df[['date', 'open', 'close', 'high', 'low', 'amount', 'code', 'name']], ignore_index = True)
        tmp_df.to_csv(data_path+"/"+file_name, index=None)
        print(file_name+" updated.")
```

```python
for df in industry_list:
    update_csv(file_map,df)
```
